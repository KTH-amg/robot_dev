```python
import pandas as pd

iris_pd = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_pd.head()
```

```python
iris_pd['species'] = iris.target
iris_pd.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(hue='species', x='sepal length (cm)', data=iris_pd, orient='h', palette='Set3')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(hue='species', x='sepal width (cm)', data=iris_pd, orient='h', palette='Set3')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(hue='species', x='petal length (cm)', data=iris_pd, orient='h', palette='Set3')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(hue='species', x='petal width (cm)', data=iris_pd, orient='h', palette='Set3')
```

```python
sns.pairplot(iris_pd, hue='species', palette='Set2')
```

```python
sns.pairplot(data=iris_pd, 
             vars=['petal length (cm)', 'petal width (cm)'],
             hue='species', palette='Set2')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)',
                data=iris_pd, hue='species', palette="Set2")
```

```python
import numpy as np

p = np.arange(0.001, 1, 0.001)
plt.plot(p, -p*np.log2(p))
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
iris_tree.fit(iris.data[:, 2:], iris.target)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(iris.data[:, 2:])
accuracy_score(iris.target, y_pred_tr)
```

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(14,8))
plot_decision_regions(X=iris.data[:, 2:], y=iris.target, clf=iris_tree, legend=2)
plt.show()
```
