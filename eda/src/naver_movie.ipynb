```python
import urllib
import urllib.parse
from urllib.request import Request
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&qvt=0&query={word1}+{word2}"
req = Request(url.format(word1=urllib.parse.quote("박스오피스"),
                         word2=urllib.parse.quote("평점순")))
page = urlopen(req)

soup = BeautifulSoup(page, "html.parser")
print(soup.prettify())
```

```python
import time

prefix = "https://search.naver.com/search.naver"
movie_panel = soup.find(class_="_panel_rating _tab_content")
movie_list = []

for idx, each in enumerate(movie_panel.find_all("li")):
    movie = []
    link = each.find("a")["href"]
    title = each.find("strong", class_="name").string
    rating = each.find("span", class_="sub_text").get_text()

    if title[-3:] == "...":
        url_link = prefix + link
        page_link = urlopen(url_link)
        soup_link = BeautifulSoup(page_link, "html.parser")
        title = soup_link.find(class_="_text").get_text()

    movie.append(str(idx + 1) + "위")
    movie.append(title)
    movie.append(rating)

    movie_list.append(movie)
    #time.sleep(0.5)

movie_list
```

```python
import pandas as pd

df = pd.DataFrame(movie_list, columns=["순위", "제목", "평점"])
df.set_index("순위", inplace=True)
df
```

```python
df.to_csv("../data/movie_ratings_standing.csv")
```
