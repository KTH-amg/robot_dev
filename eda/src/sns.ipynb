```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Seaborn 연습

```python
import numpy as np

x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)
```

```python
sns.set_style("white")
plt.figure(figsize=(10, 6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
sns.despine(offset=10) 
plt.show()
```

## Boxplot + Swarmplot

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(x="day", y="total_bill", data=tips, color=".5")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, hue="day")
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.show()
```

## Lmplot

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, height=7)
plt.show()
```

```python
tips_asc = tips.sort_values("tip", ascending=True)
tips_asc.head()
```

```python
tips_dsc = tips.sort_values("tip", ascending=False)
tips_dsc.head()
```

```python
plt.figure(figsize=(14, 10))
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)

for n in range(5):
    plt.text(tips_asc['total_bill'].iloc[n] * 1.02,
             tips_asc['tip'].iloc[n] * 0.98,
             'hmm', fontsize=15)
    
    plt.text(tips_dsc['total_bill'].iloc[n] * 1.02,
             tips_dsc['tip'].iloc[n] * 0.98,
             'wow', fontsize=15)
    
plt.grid()
plt.show()
```

## Flights 데이터(Heatmap)

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights = flights.pivot(index="month", columns="year", values="passengers")
flights.head()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
```

## Iris 데이터(pairplot)

```python
sns.set(style="ticks")
iris = sns.load_dataset("iris")
iris.head(10)
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
sns.pairplot(iris, x_vars=["sepal_width", "sepal_length"],
             y_vars=["petal_width", "petal_length"],
             hue="species",
             height=4)
plt.show()
```

## Titanic 데이터(catplot)

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.catplot(x="sex", y="survived", hue="pclass", data=titanic, kind="bar",
            palette="muted", legend=True, height=7)
plt.show()
```

```python
sns.catplot(x="deck", kind="count", hue="deck", palette="ch:.25", data=titanic)
plt.show()
```
